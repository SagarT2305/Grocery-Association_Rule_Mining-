```python
data = pd.read_csv('C:/Users/Lenovo/Downloads/Compressed/GroceryStoreDataSet.csv', header=None)
```

```python
data = data.rename(columns={0: 'items'})
data
```

```python
x = list(data['items'].apply(lambda x: x.split(',')))
x
```

# TransactionEncoder

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()

x = pd.DataFrame(te.fit_transform(x), columns=te.columns_)
x
```

# apriori

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
x1 = apriori(x, use_colnames= True, min_support=0.2)
x1
```

# association_rules

```python
x2 = association_rules(x1, min_threshold=0.5, metric='confidence')
x2
```

```python
x3= x2[['antecedents', 'consequents', 'support', 'confidence']].sort_values(by='confidence',ascending=False )
x3
```

```python
x4 = x3[x3.confidence >= 0.65]
x4
```
